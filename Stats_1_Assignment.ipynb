```python
from statistics import mean, median, mode, stdev
import numpy as np
import scipy.integrate as integrate
from scipy.stats import binom, poisson, norm
import math
import matplotlib.pyplot as plt
```

```python
# Problem Statement 1: Marks of Year 8 class
marks = [6, 7, 5, 7, 7, 8, 7, 6, 9, 7, 4, 10, 6, 8, 8, 9, 5, 6, 4, 8]

# Calculate statistics for Problem Statement 1
mean_marks = mean(marks)
median_marks = median(marks)
mode_marks = mode(marks)
stdev_marks = stdev(marks)

print("Problem Statement 1 - Marks Statistics:")
print(f"Mean: {mean_marks}")
print(f"Median: {median_marks}")
print(f"Mode: {mode_marks}")
print(f"Standard Deviation: {stdev_marks}\n")
```

```python
# Problem Statement 2: Number of calls per day for roadside service
calls = [28, 122, 217, 130, 120, 86, 80, 90, 140, 120, 70, 40, 145, 113, 90, 68,
         174, 194, 170, 100, 75, 104, 97, 75, 123, 100, 75, 104, 97, 75, 123,
         100, 89, 120, 109]

# Calculate statistics for Problem Statement 2
mean_calls = mean(calls)
median_calls = median(calls)
mode_calls = mode(calls)
stdev_calls = stdev(calls)

print("Problem Statement 2 - Calls Statistics:")
print(f"Mean: {mean_calls}")
print(f"Median: {median_calls}")
print(f"Mode: {mode_calls}")
print(f"Standard Deviation: {stdev_calls}")
```

```python
# Problem Statement 3
x = np.array([0, 1, 2, 3, 4, 5])
f_x = np.array([0.09, 0.15, 0.40, 0.25, 0.10, 0.01])

# Calculate mean
mean_workouts = np.sum(x * f_x)

# Calculate variance
variance_workouts = np.sum(((x - mean_workouts) ** 2) * f_x)

print("Problem Statement 3 - Gym Workouts:")
print(f"Mean number of workouts in a week: {mean_workouts}")
print(f"Variance: {variance_workouts}")
```

```python
# Problem Statement 4
# PDF function
def pdf(d):
    return 20 * np.exp(-20 * (d - 12.5))

# Calculate proportion of parts to be scrapped (P(D > 12.6))
P_scrap = 1 - integrate.quad(pdf, 12.5, 12.6)[0]

# Calculate CDF at D = 11 (Note: F(11) as D >= 12.5 makes it zero)
CDF_at_11 = integrate.quad(pdf, 12.5, 11)[0]

print("Problem Statement 4 - Proportion of Parts Scrapped:")
print(f"Proportion of parts to be scrapped (P(D > 12.6)): {P_scrap}")
print(f"CDF at D = 11: {CDF_at_11}")
```

```python
# Problem Statement 5
n = 6       # number of LEDs chosen
p = 0.3     # probability of a faulty LED

# Calculate probability of having exactly 2 faulty LEDs
prob_2_faulty = binom.pmf(2, n, p)

# Calculate mean and standard deviation for the binomial distribution
mean_faulty = n * p
std_dev_faulty = math.sqrt(n * p * (1 - p))

print("Problem Statement 5 - LED Faulty Rate:")
print(f"Probability of exactly 2 faulty LEDs: {prob_2_faulty}")
print(f"Mean (average) faulty LEDs in sample: {mean_faulty}")
print(f"Standard deviation: {std_dev_faulty}")
```

```python
# Problem Statement 6
# Gaurav's parameters
n_gaurav = 8
p_gaurav = 0.75

# Barakha's parameters
n_barakha = 12
p_barakha = 0.45

# Probability calculations
prob_gaurav_5 = binom.pmf(5, n_gaurav, p_gaurav)
prob_barakha_5 = binom.pmf(5, n_barakha, p_barakha)

# Probability calculations for 4 and 6 correct solutions
prob_gaurav_4 = binom.pmf(4, n_gaurav, p_gaurav)
prob_gaurav_6 = binom.pmf(6, n_gaurav, p_gaurav)
prob_barakha_4 = binom.pmf(4, n_barakha, p_barakha)
prob_barakha_6 = binom.pmf(6, n_barakha, p_barakha)

# Display results
print("Problem Statement 6 - Exam Preparation:")
print(f"Gaurav's probability of solving exactly 5 questions correctly: {prob_gaurav_5}")
print(f"Barakha's probability of solving exactly 5 questions correctly: {prob_barakha_5}")
print(f"Gaurav's probability of solving exactly 4 questions correctly: {prob_gaurav_4}")
print(f"Gaurav's probability of solving exactly 6 questions correctly: {prob_gaurav_6}")
print(f"Barakha's probability of solving exactly 4 questions correctly: {prob_barakha_4}")
print(f"Barakha's probability of solving exactly 6 questions correctly: {prob_barakha_6}")

# Plotting probabilities of correct solutions for both
x_gaurav = np.arange(0, n_gaurav + 1)
y_gaurav = binom.pmf(x_gaurav, n_gaurav, p_gaurav)
x_barakha = np.arange(0, n_barakha + 1)
y_barakha = binom.pmf(x_barakha, n_barakha, p_barakha)

plt.figure(figsize=(12, 5))

# Gaurav's plot
plt.subplot(1, 2, 1)
plt.bar(x_gaurav, y_gaurav, color='blue', alpha=0.6)
plt.title("Probability Distribution - Gaurav")
plt.xlabel("Number of Correct Solutions")
plt.ylabel("Probability")

# Barakha's plot
plt.subplot(1, 2, 2)
plt.bar(x_barakha, y_barakha, color='orange', alpha=0.6)
plt.title("Probability Distribution - Barakha")
plt.xlabel("Number of Correct Solutions")
plt.ylabel("Probability")

plt.tight_layout()
plt.show()
```

```python
# Problem Statement 7
# Parameters for Poisson distribution
lambda_rate = 4.8  # arrival rate in 4 minutes

# Part a) Probability of exactly 5 customers
prob_5_customers = poisson.pmf(5, lambda_rate)

# Part b) Probability of not more than 3 customers (0, 1, 2, or 3 customers)
prob_not_more_than_3 = poisson.cdf(3, lambda_rate)

# Part c) Probability of more than 3 customers
prob_more_than_3 = 1 - prob_not_more_than_3

# Display results
print("Problem Statement 7 - Customer Arrivals:")
print(f"Probability of exactly 5 customers arriving: {prob_5_customers}")
print(f"Probability of not more than 3 customers: {prob_not_more_than_3}")
print(f"Probability of more than 3 customers: {prob_more_than_3}")

# Visualize with a bar plot
x_values = np.arange(0, 15)  # reasonable range for number of customers
y_values = poisson.pmf(x_values, lambda_rate)

plt.figure(figsize=(10, 5))
plt.bar(x_values, y_values, color='purple', alpha=0.6)
plt.title("Poisson Distribution of Customer Arrivals (λ = 4.8)")
plt.xlabel("Number of Customers")
plt.ylabel("Probability")
plt.show()
```

```python
# Problem Statement 8
errors_per_hour = 6          # errors per hour
words_per_minute = 77        # words per minute

# Calculate the error rate per word and apply to different report lengths
error_rate_per_word = errors_per_hour / (words_per_minute * 60)

# Lambda for different word counts
lambda_455 = error_rate_per_word * 455
lambda_1000 = error_rate_per_word * 1000
lambda_255 = error_rate_per_word * 255

# Probability of exactly 2 errors for 455-word report
prob_2_errors_455 = poisson.pmf(2, lambda_455)

# Display results for 455 words
print("Problem Statement 8 - Error Rate in Report Writing:")
print(f"Probability of exactly 2 errors in a 455-word report: {prob_2_errors_455}")

# Effects when words increase or decrease
prob_2_errors_1000 = poisson.pmf(2, lambda_1000)
prob_2_errors_255 = poisson.pmf(2, lambda_255)

print(f"Probability of exactly 2 errors in a 1000-word report: {prob_2_errors_1000}")
print(f"Probability of exactly 2 errors in a 255-word report: {prob_2_errors_255}")

# Visualize how λ affects PMF
word_counts = [255, 455, 1000]
lambdas = [lambda_255, lambda_455, lambda_1000]
error_counts = np.arange(0, 10)

plt.figure(figsize=(12, 6))
for i, lam in enumerate(lambdas):
    pmf_values = poisson.pmf(error_counts, lam)
    plt.plot(error_counts, pmf_values, marker='o', linestyle='-', label=f"{word_counts[i]} words")

plt.title("PMF of Errors for Different Word Counts")
plt.xlabel("Number of Errors")
plt.ylabel("Probability")
plt.legend(title="Word Count")
plt.show()
```

```python
# Problem Statement 9
# Parameters
lambda_rate = 20  # rate parameter for the exponential distribution
d_target = 12.5  # target diameter
d_scrap = 12.6   # scrap diameter threshold

# 1. Proportion of parts with diameter > 12.6 mm
prob_scrap = np.exp(-lambda_rate * (d_scrap - d_target))

# 2. CDF when diameter is 11 mm
# Since 11 is outside the defined range, we can set it directly to 0
cdf_at_11 = 0

# Display results
print("Problem Statement 9 - Proportion and CDF Calculations:")
print(f"Proportion of parts with diameter > 12.6 mm (scrap rate): {prob_scrap}")
print(f"CDF when diameter is 11 mm: {cdf_at_11}")
```

```python
# Problem Statement 10

# 1. Probability calculations using standard normal distribution
p_gt_1_26 = 1 - norm.cdf(1.26)
p_lt_neg_0_86 = norm.cdf(-0.86)
p_gt_neg_1_37 = 1 - norm.cdf(-1.37)
p_between_neg1_25_0_37 = norm.cdf(0.37) - norm.cdf(-1.25)
p_leq_neg_4_6 = norm.cdf(-4.6)

# 2. Find z values for given probabilities
z_gt_0_05 = norm.ppf(1 - 0.05)
z_between_0_99 = norm.ppf(0.995)  # 0.995 for one side in symmetric interval

# Display results
print("Problem Statement 10 - Standard Normal Probabilities and Quantiles:")
print(f"P(Z > 1.26): {p_gt_1_26}")
print(f"P(Z < -0.86): {p_lt_neg_0_86}")
print(f"P(Z > -1.37): {p_gt_neg_1_37}")
print(f"P(-1.25 < Z < 0.37): {p_between_neg1_25_0_37}")
print(f"P(Z ≤ -4.6): {p_leq_neg_4_6}")
print(f"z such that P(Z > z) = 0.05: {z_gt_0_05}")
print(f"z such that P(-z < Z < z) = 0.99: ±{z_between_0_99}")
```

```python
# Problem Statement 11
mean_current = 10  # mean (mA)
std_dev_current = 2  # standard deviation (mA)

# 1. Probability that current measurement exceeds 13 mA
z_13 = (13 - mean_current) / std_dev_current
p_exceeds_13 = 1 - norm.cdf(z_13)

# 2. Probability that current measurement is between 9 and 11 mA
z_9 = (9 - mean_current) / std_dev_current
z_11 = (11 - mean_current) / std_dev_current
p_between_9_11 = norm.cdf(z_11) - norm.cdf(z_9)

# 3. Current measurement for cumulative probability of 0.98
z_0_98 = norm.ppf(0.98)
x_0_98 = mean_current + z_0_98 * std_dev_current

# Display results
print("Problem Statement 11 - Current Measurement Probabilities:")
print(f"Probability that current exceeds 13 mA: {p_exceeds_13}")
print(f"Probability that current is between 9 and 11 mA: {p_between_9_11}")
print(f"Current measurement corresponding to 0.98 probability: {x_0_98} mA")
```

```python
# Problem Statement 12
mean_diameter_current = 0.2508  # mean diameter (inches)
std_dev_diameter = 0.0005  # standard deviation (inches)
spec_min = 0.2485  # lower spec limit
spec_max = 0.2515  # upper spec limit

# 1. Proportion of shafts within specifications (current mean)
z_spec_min_current = (spec_min - mean_diameter_current) / std_dev_diameter
z_spec_max_current = (spec_max - mean_diameter_current) / std_dev_diameter
p_within_specs_current = norm.cdf(z_spec_max_current) - norm.cdf(z_spec_min_current)

# 2. Proportion of shafts within specifications (target mean: 0.2500)
mean_diameter_target = 0.2500  # target mean diameter
z_spec_min_target = (spec_min - mean_diameter_target) / std_dev_diameter
z_spec_max_target = (spec_max - mean_diameter_target) / std_dev_diameter
p_within_specs_target = norm.cdf(z_spec_max_target) - norm.cdf(z_spec_min_target)

# Display results
print("Problem Statement 12 - Shaft Diameter Proportion:")
print(f"Proportion within specifications (current mean of 0.2508): {p_within_specs_current}")
print(f"Proportion within specifications (target mean of 0.2500): {p_within_specs_target}")
```

# **Conclusion**
The proportion of shafts within specifications gives an indication of production quality. If the proportion is low, many shafts do not meet the required tolerance.
Adjusting the mean to the target value increases the likelihood of meeting specifications, which may reduce manufacturing defects or rework.
By centering the process, the production can be optimized to meet quality standards consistently.
